```python
% cd /Users/elliedobson/Documents/Work/Customers/TRG/Code
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML
import Utils.PythonUtils.TRGPythonFunctions
from Utils.PythonUtils.BaseWidgetController import BaseWidgetController
import matplotlib.pyplot as plt
% matplotlib inline

class NewInWidgetController(BaseWidgetController):
    
    def __init__(self):
        BaseWidgetController.__init__(self)
        
        # set additional parameters
        self.Table = 'newin_master'
        self.ExtraCond =''
        self.Brand=''
        self.MinPrice=0
        self.MaxPrice=5000
    
    def SetupUI(self): 
        # set country
        self.CountryWidget = widgets.RadioButtonsWidget(values=['uk', 'us', 'au'], description = "Country:")
        display(self.CountryWidget)
        
        # set analysis option
        self.AnalysisWidget = widgets.RadioButtonsWidget(values=['', 'Declining subcategories', 'Increasing subcategories'], description = "Choose analysis:")
        self.AnalysisWidget.on_trait_change(self.GetCategories, "value")
        display(self.AnalysisWidget)
        
    def GetCategories(self):         
        self.UpdateValues()
        BaseWidgetController.GetCategories(self)
        self.CategoryWidget.on_trait_change(self.GetSubCategories, "value")
        
    def GetSubCategories(self):
        self.UpdateValues()
        BaseWidgetController.GetSubCategories(self)
        self.SubCategoryWidget.on_trait_change(self.GetRetailers, "value")
        
    def GetRetailers(self):
        self.UpdateValues()
        Retailers=self.fe.executeandfetch("select retailername from " + self.Table + \
                                          " where category = '" + self.Category + \
                                          "' and subcategory like '%" + self.SubCategory + \
                                          "%' and facetbrand like '%" + self.Brand + \
                                          "%' and country = '" + self.Country + \
                                          "'" + self.ExtraCond + \
                                          " and newin_7 = 1 " + \
                                          " group by 1 order by 1");
        Retailers=self.PrepareList(Retailers, ["All", ""])
        if len(Retailers)==0:
                print "Not enough data. Choose another subcategory"
        self.RetailerWidget = widgets.DropdownWidget(values=Retailers, description = "Choose a retailer: ")
        display(self.RetailerWidget) 
        self.CreateGo()
        
    def GetVals(self, value):
        self.UpdateValues()
        BatchStatement1 = "select batch_date, count(*) from " + self.Table + \
                          " where country = '" + self.Country + \
                          "' and category = '" + self.Category + \
                          "' and subcategory like '%" + self.SubCategory + \
                          "%' and retailername like '%" + self.Retailer + \
                          "%'" + self.ExtraCond + \
                          " group by 1 order by 1"
                        
        BatchStatement2 = "select batch_date, count(*) from " + self.Table + \
                          " where country = '" + self.Country + \
                          "' and category = '" + self.Category + \
                          "' and subcategory like '%" + self.SubCategory + \
                          "%' and retailername like '%" + self.Retailer + \
                          "%'" + self.ExtraCond + \
                          " and newin_7 = 1 " + \
                          " group by 1 order by 1"

        print BatchStatement1        
        print BatchStatement2
        
        Mix_all=self.fe.executeandfetch(BatchStatement1)
        Mix_all=zip(*Mix_all)
        self.x_all=Mix_all[0]
        Total=sum(Mix_all[1])
        self.y_all=[100*float(x)/Total for x in Mix_all[1]]
        
        Mix_newin=self.fe.executeandfetch(BatchStatement2)
        Mix_newin=zip(*Mix_newin)
        self.x_newin=Mix_newin[0]
        Total=sum(Mix_newin[1])
        self.y_newin=[100*float(x)/Total for x in Mix_newin[1]]
        
        self.MakePlot()
        
    def MakePlot(self):
        self.UpdateValues()

        plt.figure(figsize=(40, 10))
        plt.tick_params(axis='both', labelsize=20)
        plt.plot_date(self.x_newin, self.y_newin, 'bo', linestyle='dashed', label = "New in last 7 days only")
        plt.plot_date(self.x_all, self.y_all, 'ro', linestyle='dashed', label = "All products")

        plt.xlabel("Batch processing date", fontsize=40)
        plt.ylabel("% mix of total", fontsize=40)
        plt.title("Mix of (" + str(self.Category) + \
                  " : " + str(self.SubCategory) + \
                  ") over time for " + self.Label, fontsize=40)
        plt.legend(loc='upper right', fontsize=40)
        plt.show()
                
    def UpdateValues(self):
        BaseWidgetController.UpdateValues(self)
        if self.AnalysisType=="Declining subcategories":
            self.ExtraCond=" and fullprice=0"
            self.Label="All marked down products"
        elif self.AnalysisType=="Increasing subcategories":
            self.ExtraCond=" and fullprice=1"
            self.Label="All full price products"
        self.Brand=''
```
